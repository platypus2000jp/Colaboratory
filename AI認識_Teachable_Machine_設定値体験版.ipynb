```python
# prompt: AITeachable MachinePython

# 
import tensorflow as tf
import numpy as np
from ipywidgets import interact, FloatSlider, IntSlider
import matplotlib.pyplot as plt

#  ()
# 2 (01) 
num_samples = 100
X_train = np.random.rand(num_samples, 2)  # 2
y_train = np.random.randint(0, 2, num_samples)  # 

# 
model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(10, activation='relu', input_shape=(2,)),
  tf.keras.layers.Dense(1, activation='sigmoid')
])

# 
def train_model(epochs, batch_size, learning_rate):
  optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
  model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
  history = model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=0)

  # 
  plt.figure(figsize=(10, 4))
  plt.subplot(1, 2, 1)
  plt.plot(history.history['loss'])
  plt.title('Loss')
  plt.xlabel('Epoch')
  plt.ylabel('Loss')

  plt.subplot(1, 2, 2)
  plt.plot(history.history['accuracy'])
  plt.title('Accuracy')
  plt.xlabel('Epoch')
  plt.ylabel('Accuracy')
  plt.show()


# 
interact(train_model,
         epochs=IntSlider(min=1, max=100, step=1, value=10),
         batch_size=IntSlider(min=1, max=32, step=1, value=16),
         learning_rate=FloatSlider(min=0.0001, max=0.1, step=0.0001, value=0.001));
```
