```python
# prompt: 画像をアップロードします。元画像と「P型（1型）色覚」シミュレーションを並べて表示してください。依存関係をインストールしてください。

!pip install opencv-python matplotlib

import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

def simulate_protanopia(image):
  """
  P型色覚シミュレーション

  Args:
    image: OpenCV画像

  Returns:
    P型色覚シミュレーション画像
  """
  rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  protanopia_matrix = np.array([[0.567, 0.433, 0.000],
                               [0.558, 0.442, 0.000],
                               [0.000, 0.242, 0.758]])

  simulated_image = np.dot(rgb_image, protanopia_matrix.T)
  simulated_image = np.clip(simulated_image, 0, 255).astype(np.uint8)
  return simulated_image

# 画像をアップロード
from google.colab import files
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # 画像読み込み
  image = cv2.imread(fn)

  # P型色覚シミュレーション
  simulated_image = simulate_protanopia(image)

  # 元画像とシミュレーション画像を並べて表示
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.title("Original Image")
  imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.subplot(1, 2, 2)
  plt.title("Protanopia Simulation")
  imshow(simulated_image)
  plt.show()
```

```python
# prompt: 画像をアップロードします。元画像と「D型（2型）色覚」のシミュレーションを並べて表示してください。依存関係をインストールしてください。

import cv2
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from google.colab import files

!pip install opencv-python matplotlib


def simulate_deuteranopia(image):
  """
  D型（2型）色覚シミュレーション

  Args:
    image: OpenCV画像

  Returns:
    D型色覚シミュレーション画像
  """
  rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  deuteranopia_matrix = np.array([[0.625, 0.375, 0.000],
                                 [0.700, 0.300, 0.000],
                                 [0.000, 0.300, 0.700]])

  simulated_image = np.dot(rgb_image, deuteranopia_matrix.T)
  simulated_image = np.clip(simulated_image, 0, 255).astype(np.uint8)
  return simulated_image

# 画像をアップロード
uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # 画像読み込み
  image = cv2.imread(fn)

  # D型色覚シミュレーション
  simulated_image = simulate_deuteranopia(image)

  # 元画像とシミュレーション画像を並べて表示
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.title("Original Image")
  imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
  plt.subplot(1, 2, 2)
  plt.title("Deuteranopia Simulation")
  imshow(simulated_image)
  plt.show()
```
