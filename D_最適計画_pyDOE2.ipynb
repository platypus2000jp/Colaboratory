```python
# prompt: D-最適計画を使用して実験を計画してください。plotlyを使用して試験表を作成してください。対象は紙飛行機とします。試験数は何回ですか。pyDOE2を使用してください。

!pip install pyDOE2
!pip install plotly

import numpy as np
import pandas as pd
from pyDOE2 import *
import plotly.graph_objects as go

# D最適計画による実験計画
# 因子数
n_factors = 4

# 実験回数
n_experiments = (n_factors + 1)


# 各因子のレベル数
levels = [5, 5, 5, 5]

# D最適計画の実行
design = lhs(n_factors, samples=n_experiments, criterion='center')


# 各因子の範囲を設定
factor_ranges = [
    [0.1, 0.3],  # 翼幅
    [0.05, 0.15],  # 翼弦長
    [15, 30],  # 翼面積
    [0, 90],  # 翼面形状
]

# 実験計画表の作成
experimental_design = pd.DataFrame(design, columns=['翼幅', '翼弦長', '翼面積', '翼面形状'])

# 各因子に範囲を適用
for i, factor_name in enumerate(experimental_design.columns):
  experimental_design[factor_name] = experimental_design[factor_name] * (factor_ranges[i][1] - factor_ranges[i][0]) + factor_ranges[i][0]


# plotlyを使用して試験表を作成
fig = go.Figure(data=[go.Table(
    header=dict(values=list(experimental_design.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[experimental_design[col] for col in experimental_design.columns],
               fill_color='lavender',
               align='left'))
])

fig.update_layout(title='紙飛行機実験計画表(D最適計画)')
fig.show()

print(f'試験回数: {n_experiments}')
```
