## 📊 Google Colabで株価ダッシュボードを作成する基本ステップ

基本的なダッシュボードを作成するためのステップは以下の通りです。

-----

### 1\. 必要なライブラリのインストールとインポート

Google Colabのノートブックで、まず必要なライブラリをインストールし、インポートします。**yfinance**はYahoo\! Financeから株価データを取得するために広く使われています。

Gemini
https://gemini.google.com/share/3bbe17639d36

```
# yfinance, pandas, matplotlib, plotlyなどをインストール
!pip install yfinance pandas matplotlib plotly --quiet

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import datetime
```

-----

### 2\. NASDAQの銘柄コードの定義とデータ取得

ダッシュボードに表示したい**NASDAQ**に上場している銘柄（例：Apple, Microsoft, NASDAQ 100 Index）の**ティッカーシンボル**を定義し、データを取得します。

  * **Apple:** `AAPL`
  * **Microsoft:** `MSFT`
  * **NASDAQ 100 Index:** `^NDX` (インデックスの場合は、ティッカーの前に `^` がつくことが多いです)

ここでは、複数の銘柄のデータを取得する例を示します。

```
# 分析したいNASDAQのティッカーシンボルをリストで定義
tickers = ["AAPL", "MSFT", "^NDX"]

# データの期間を定義（例：過去1年間）
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=365)

# yfinanceを使って株価データを取得
data = yf.download(tickers, start=start_date, end=end_date)
```

-----

### 3\. データの前処理と表示

取得したデータは通常、マルチインデックス（複数の列グループ）になっています。ここでは、ダッシュボードによく使われる**終値** (`Close`) のみを抽出します。

```
# 終値 (Close) のみを抽出
closing_prices = data['Close']

# 最初の数行を表示して確認
print(closing_prices.head())
```

-----

### 4\. 株価チャートの作成（データ可視化）

データを視覚化することで、株価のトレンドを把握できます。**Plotly**を使うと、インタラクティブ（マウスで操作できる）なチャートを簡単に作成できます。

```
# Plotly Expressを使って株価のトレンドをプロット
fig = px.line(
    closing_prices,
    title='NASDAQ 主要銘柄の株価推移',
    labels={'value': '株価 (USD)', 'Date': '日付'}
)

# チャートの表示
fig.show()
```

### 📈 さらに高度なダッシュボード機能

上記の基本ステップに加えて、以下の機能を追加することで、より実用的なダッシュボードを作成できます。

  * **テクニカル指標の追加:** **移動平均線 (SMA/EMA)** や **RSI** などを計算し、チャートに重ねて表示する。
  * **ローソク足チャート:** `plotly.graph_objects` を使用して、始値・高値・安値・終値を含むローソク足チャートを作成する。
  * **インタラクティブな入力:** Colabのフォーム機能 (`#@param` コメント) を使用して、分析期間や銘柄をユーザーが簡単に入力できるようにする。
  * **株価変動率の計算:** 日次または期間全体の**リターン**を計算し、比較する。
