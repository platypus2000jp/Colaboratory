```python
# prompt: アップロードしたPDFファイルのファイル名、ページ数、作成日、最終変更日、印刷日時、アクセス日を表にしてください。
# pypdfとPlotlyを使ってください。

!pip install pypdf plotly

from pypdf import PdfReader
import plotly.graph_objects as go
import pandas as pd

uploaded_file_name = list(uploaded.keys())[0]

reader = PdfReader(uploaded_file_name)
number_of_pages = len(reader.pages)

# Get metadata
metadata = reader.metadata
creation_date = metadata.get('/CreationDate', 'N/A')
modification_date = metadata.get('/ModDate', 'N/A')
print_date = metadata.get('/PrintDate', 'N/A')  # May not always be available
access_date = metadata.get('/AccessDate', 'N/A')  # May not always be available


data = {
    'ファイル名': [uploaded_file_name],
    'ページ数': [number_of_pages],
    '作成日': [creation_date],
    '最終変更日': [modification_date],
    '印刷日時': [print_date],
    'アクセス日': [access_date],
}

df = pd.DataFrame(data)


fig = go.Figure(data=[go.Table(
    header=dict(values=list(df.columns),
                fill_color='paleturquoise',
                align='left'),
    cells=dict(values=[df[col] for col in df.columns],
               fill_color='lavender',
               align='left'))
])

fig.update_layout(title='PDFファイル情報')
fig.show()
```

```python
# prompt: PDFに強力な暗号化を行うことはできますか。pypdfを使ってください。

from PyPDF2 import PdfWriter, PdfReader

def encrypt_pdf(input_pdf_path, output_pdf_path, user_password, owner_password=None):
  """
  PDFファイルを暗号化します。

  Args:
    input_pdf_path: 入力PDFファイルのパス。
    output_pdf_path: 出力PDFファイルのパス。
    user_password: ユーザーパスワード。
    owner_password: オーナーパスワード（オプション）。
  """
  try:
    with open(input_pdf_path, 'rb') as f:
      reader = PdfReader(f)
      writer = PdfWriter()

      for page in reader.pages:
        writer.add_page(page)

      writer.encrypt(user_password=user_password, owner_password=owner_password, use_128bit=True)

      with open(output_pdf_path, 'wb') as output_file:
        writer.write(output_file)

      print(f"PDFファイルが暗号化されました: {output_pdf_path}")

  except Exception as e:
    print(f"エラーが発生しました: {e}")


# 使用例
input_pdf_path = 'input.pdf'  # 入力PDFファイルのパス
output_pdf_path = 'encrypted_output.pdf'  # 出力PDFファイルのパス
user_password = 'my_password'  # ユーザーパスワード
owner_password = 'my_owner_password'  # オーナーパスワード（オプション）

encrypt_pdf(input_pdf_path, output_pdf_path, user_password, owner_password)
```
