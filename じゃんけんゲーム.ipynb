```python
# prompt: じゃんけんゲームを作ってください。プレーヤーはグー・チョキ・パーのボタンから一つ選びます。勝ち負けを表示してください。もう一回のボタンでもう一度開始します。

import random
import ipywidgets as widgets
from IPython.display import display

def janken(player_choice):
  choices = ['グー', 'チョキ', 'パー']
  computer_choice = random.choice(choices)

  print(f"あなたは{player_choice}を選びました。")
  print(f"コンピューターは{computer_choice}を選びました。")

  if player_choice == computer_choice:
    print("引き分けです！")
  elif (player_choice == 'グー' and computer_choice == 'チョキ') or \
       (player_choice == 'チョキ' and computer_choice == 'パー') or \
       (player_choice == 'パー' and computer_choice == 'グー'):
    print("あなたの勝ちです！")
  else:
    print("コンピューターの勝ちです！")

  restart_button.disabled = False


# ボタンの作成
gu_button = widgets.Button(description='グー')
choki_button = widgets.Button(description='チョキ')
pa_button = widgets.Button(description='パー')
restart_button = widgets.Button(description='もう一回')
restart_button.disabled = True

# ボタンがクリックされたときの処理
def on_gu_button_clicked(b):
  janken('グー')

def on_choki_button_clicked(b):
  janken('チョキ')

def on_pa_button_clicked(b):
  janken('パー')

def on_restart_button_clicked(b):
  restart_button.disabled = True
  print("\nじゃんけん開始！")


gu_button.on_click(on_gu_button_clicked)
choki_button.on_click(on_choki_button_clicked)
pa_button.on_click(on_pa_button_clicked)
restart_button.on_click(on_restart_button_clicked)

# ボタンを表示
print("じゃんけん開始！")
display(widgets.HBox([gu_button, choki_button, pa_button]))
display(restart_button)
```
