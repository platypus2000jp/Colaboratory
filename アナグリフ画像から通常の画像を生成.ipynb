```python
# prompt: アナグリフ画像をアップロードします。アナグリフ画像から通常の画像を生成してください。左にアナグリフ画像を表示し、右に生成した通常画像を並べます。
# 処理後の通常画像を名前を付けて保存しダウンロードします。

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# アナグリフ画像をアップロード
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# 画像読み込み
img = cv2.imread(filename)

# アナグリフ画像を左右の画像に分割
width = img.shape[1]
left_img = img[:, :width // 2, :]
right_img = img[:, width // 2:, :]


# 左右の画像を合成して通常画像を生成
normal_img = cv2.addWeighted(left_img, 0.5, right_img, 0.5, 0)

# 画像の表示
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
axes[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0].set_title('Anaglyph Image')
axes[1].imshow(cv2.cvtColor(normal_img, cv2.COLOR_BGR2RGB))
axes[1].set_title('Generated Normal Image')
plt.show()


# 処理後の画像を保存
output_filename = 'normal_image.jpg'
cv2.imwrite(output_filename, normal_img)

# ダウンロード
files.download(output_filename)
```
