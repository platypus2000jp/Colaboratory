```python
# prompt: フィッシャーのアヤメのテーブルを見せてください。

from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target
print(df)
```

```python
# prompt: アヤメをSweetvizでレポートにしてください。

!pip install sweetviz

import sweetviz as sv

report = sv.analyze(df)
report.show_html()
```

```python
# prompt: アヤメを全数プロットとバイオリンプロットで可視化してください。

import matplotlib.pyplot as plt
import seaborn as sns

# 全数プロット
sns.pairplot(df, hue='target')
plt.show()

# バイオリンプロット
for col in iris.feature_names:
  plt.figure(figsize=(8, 6))
  sns.violinplot(x='target', y=col, data=df)
  plt.title(f'Violin Plot of {col}')
  plt.show()
```

```python
# prompt: アヤメをレインドロップ・プロットで可視化してください。

import seaborn as sns
import matplotlib.pyplot as plt

# レインドロッププロットを作成
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=df, kind='kde')
plt.show()

sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=df, kind='kde')
plt.show()
```

```python
# prompt: アヤメをビーン・プロットで可視化してください。

import seaborn as sns
import matplotlib.pyplot as plt

# Bean plot を作成
for col in iris.feature_names:
  plt.figure(figsize=(8, 6))
  sns.violinplot(x='target', y=col, data=df)
  sns.swarmplot(x='target', y=col, data=df, color='black')
  plt.title(f'Bean Plot of {col}')
  plt.show()
```

```python
# prompt: アヤメをビー・スウォーム・プロットで可視化してください。

import seaborn as sns
import matplotlib.pyplot as plt

# Bean plot を作成
for col in iris.feature_names:
  plt.figure(figsize=(8, 6))
  sns.violinplot(x='target', y=col, data=df)
  sns.swarmplot(x='target', y=col, data=df, color='black')
  plt.title(f'Bean Plot of {col}')
  plt.show()
```
