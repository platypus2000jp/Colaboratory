```python
# prompt: 画像のコントラストをチェックするチェッカーを作ることはできますか。画像をアップロードして、元画像とコントラストチェックを並べて表示します。

from google.colab import files
import cv2
import numpy as np
from matplotlib import pyplot as plt

def check_contrast(image_path):
  """画像のコントラストをチェックし、元画像とコントラストチェックを並べて表示します。

  Args:
    image_path: 画像のファイルパス。
  """

  # 画像を読み込む
  img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

  # コントラストを計算
  mean = np.mean(img)
  std = np.std(img)
  contrast = std / mean

  # コントラストチェックを行う
  # ここでは、単純にヒストグラム平坦化でコントラストを調整しています。
  # より高度なコントラスト調整方法を使用することも可能です。
  equ = cv2.equalizeHist(img)

  # 元画像とコントラストチェックを並べて表示
  plt.figure(figsize=(10, 5))
  plt.subplot(1, 2, 1)
  plt.imshow(img, cmap='gray')
  plt.title('Original Image')
  plt.subplot(1, 2, 2)
  plt.imshow(equ, cmap='gray')
  plt.title('Contrast Checked Image')
  plt.show()

  print(f"コントラスト: {contrast:.2f}")

# ファイルをアップロード
uploaded = files.upload()

# アップロードされた画像のファイル名を取得
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

  # コントラストチェックを実行
  check_contrast(fn)
```
