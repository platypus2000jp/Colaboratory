```python
# prompt: フラッシュ暗算アプリを作ってください。もう一度挑戦しますか？の質問は不要です。次々問題を出してください。挑戦回数と正解回数を表示してください。計算はランダムな四則演算としてください。

import random

def generate_problem():
  """ランダムな四則演算の問題を生成する"""
  num1 = random.randint(1, 10)
  num2 = random.randint(1, 10)
  operator = random.choice(['+', '-', '*', '/'])
  problem = f"{num1} {operator} {num2}"
  return problem, eval(problem)

def main():
  """フラッシュ暗算アプリのメイン関数"""
  challenge_count = 0
  correct_count = 0

  while True:
    problem, answer = generate_problem()
    print(f"問題: {problem}")

    try:
      user_answer = float(input("答えを入力してください: "))
    except ValueError:
      print("無効な入力です。数値を入力してください。")
      continue

    challenge_count += 1
    if abs(user_answer - answer) < 1e-6:
      print("正解！")
      correct_count += 1
    else:
      print(f"不正解です。正解は {answer} です。")

    print(f"挑戦回数: {challenge_count}, 正解回数: {correct_count}")

if __name__ == "__main__":
  main()
```
