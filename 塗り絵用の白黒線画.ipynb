```python
# prompt: 画像をアップロードするので、塗り絵用の白黒線画を生成してください。
# 左に元画像を、右に処理後の塗り絵用の白黒線画を並べて表示します。
# 処理後の塗り絵用の白黒線画に名前を付けて保存しダウンロードします。
# 以下のコードを適用します。
# import cv2
# import numpy as np
# def generate_line_art(image_path, output_path):
#     # 画像を読み込む
#     image = cv2.imread(image_path)
#     # グレースケールに変換
#     gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#     # 画像を反転
#     inverted_image = 255 - gray_image
#     # 反転した画像をぼかす
#     blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
#     # 元のグレースケール画像とぼかした画像を合成
#     sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=256)
#     # 合成した画像を保存
#     cv2.imwrite(output_path, sketch_image)
#     print(f"線画が保存されました: {output_path}")
# # 使用例
# image_path = 'input_image.jpg'  # 元の画像ファイルパス
# output_path = 'output_line_art.jpg'  # 出力する線画画像ファイルパス
# generate_line_art(image_path, output_path)

import cv2
import numpy as np
from google.colab import files
from google.colab.patches import cv2_imshow

def generate_line_art(image_path, output_path):
    # 画像を読み込む
    image = cv2.imread(image_path)
    # グレースケールに変換
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # 画像を反転
    inverted_image = 255 - gray_image
    # 反転した画像をぼかす
    blurred_image = cv2.GaussianBlur(inverted_image, (21, 21), 0)
    # 元のグレースケール画像とぼかした画像を合成
    sketch_image = cv2.divide(gray_image, 255 - blurred_image, scale=256)
    # 合成した画像を保存
    cv2.imwrite(output_path, sketch_image)
    print(f"線画が保存されました: {output_path}")

    # 画像を表示
    # 元画像と線画を横に並べて表示
    combined_image = np.concatenate((image, cv2.cvtColor(sketch_image, cv2.COLOR_GRAY2BGR)), axis=1)
    cv2_imshow(combined_image)

    # ダウンロード
    files.download(output_path)


# アップロードされた画像のファイル名を取得
uploaded = files.upload()
image_name = list(uploaded.keys())[0]

# 使用例
image_path = image_name  # 元の画像ファイルパス
output_path = 'output_line_art.jpg'  # 出力する線画画像ファイルパス
generate_line_art(image_path, output_path)
```
