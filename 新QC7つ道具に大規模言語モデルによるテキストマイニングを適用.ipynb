```python
# prompt: 不具合記録は文章で記録します。親和図法を扱うライブラリがあれば例題を見せてください。新QC7つ道具の言語データの処理には大規模言語モデル（LLM）を使用して文章のテキストマイニングを行ってください。plotlyで作図・作表してください。

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from collections import defaultdict

# 大規模言語モデル（LLM）の機能を模倣した簡易的なテキストマイニング関数
def analyze_text(text):
  """
  簡易的なテキストマイニング関数
  LLMのように文章からキーワードや関連性を抽出することを想定
  """
  keywords = []
  if "キズ" in text:
    keywords.append("キズ")
  if "汚れ" in text:
    keywords.append("汚れ")
  if "変形" in text:
    keywords.append("変形")
  return keywords


# 不具合記録（例）
bugu_records = [
    {"工程": "A", "不良品の種類": "キズ", "発生数": 5, "詳細": "部品の表面に深いキズが見られる"},
    {"工程": "A", "不良品の種類": "汚れ", "発生数": 3, "詳細": "部品に油汚れが付着している"},
    {"工程": "B", "不良品の種類": "キズ", "発生数": 2, "詳細": "部品の角に小さなキズがある"},
    {"工程": "B", "不良品の種類": "変形", "発生数": 1, "詳細": "部品がわずかに変形している"},
    {"工程": "C", "不良品の種類": "汚れ", "発生数": 4, "詳細": "部品に粉塵が付着している"},
    {"工程": "C", "不良品の種類": "変形", "発生数": 2, "詳細": "部品が大きく変形している"},
]

# DataFrameに変換
df_bugu = pd.DataFrame(bugu_records)

# 親和図法のような処理（簡易版）
relation_matrix = defaultdict(lambda: defaultdict(int))
for index, row in df_bugu.iterrows():
  keywords = analyze_text(row["詳細"])
  for i in range(len(keywords)):
    for j in range(i + 1, len(keywords)):
      relation_matrix[keywords[i]][keywords[j]] += 1

# 親和図法の結果をDataFrameに変換
relation_df = pd.DataFrame(relation_matrix)

# Plotlyでヒートマップを作成
fig_heatmap = px.imshow(relation_df, labels=dict(x="不良要因", y="不良要因", color="関連度"),
                       title="不良要因間の関連度")
fig_heatmap.show()

# 工程別の不良発生数の棒グラフ
fig_bar_process = px.bar(df_bugu.groupby("工程")["発生数"].sum(),
                        title="工程別の不良発生数")
fig_bar_process.show()

# 不良品の種類別の発生数の棒グラフ
fig_bar_type = px.bar(df_bugu.groupby("不良品の種類")["発生数"].sum(),
                      title="不良品の種類別の発生数")
fig_bar_type.show()

# パレート図
df_pareto = df_bugu.groupby("不良品の種類")["発生数"].sum().sort_values(ascending=False)
fig_pareto = go.Figure()
fig_pareto.add_trace(go.Bar(x=df_pareto.index, y=df_pareto.values))
fig_pareto.add_trace(go.Scatter(x=df_pareto.index, y=df_pareto.cumsum().values, mode='lines', name='累積割合'))
fig_pareto.update_layout(title="パレート図")
fig_pareto.show()
```

```python
# prompt: 不具合記録は文章で記録します。アローダイヤグラム法を扱うライブラリがあれば例題を見せてください。新QC7つ道具の言語データの処理には大規模言語モデル（LLM）を使用して文章のテキストマイニングを行ってください。plotlyで作図・作表してください。

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from collections import defaultdict

# ... (既存のコードは省略) ...


# アローダイヤグラム法の簡易的な実装（例）
def create_arrow_diagram(df_bugu):
    """
    アローダイヤグラムを作成する関数
    """
    process_list = sorted(list(set(df_bugu['工程'])))
    problem_list = sorted(list(set(df_bugu['不良品の種類'])))

    # ノード間の関係性を表す辞書
    relation_dict = {}
    for process in process_list:
        relation_dict[process] = {}
        for problem in problem_list:
            relation_dict[process][problem] = 0

    for index, row in df_bugu.iterrows():
        relation_dict[row['工程']][row['不良品の種類']] += row['発生数']

    # グラフの作成
    fig_arrow = go.Figure()

    # 工程ノード
    for process in process_list:
        fig_arrow.add_shape(
            type='circle',
            xref='x', yref='y',
            x0=process_list.index(process) - 0.2, y0=0.5,
            x1=process_list.index(process) + 0.2, y1=1.5,
            line_color='black'
        )
        fig_arrow.add_annotation(
            x=process_list.index(process), y=1,
            text=process,
            showarrow=False
        )

    # 不良品の種類ノード
    for problem in problem_list:
        fig_arrow.add_shape(
            type='circle',
            xref='x', yref='y',
            x0=len(process_list) + problem_list.index(problem) - 0.2, y0=0.5,
            x1=len(process_list) + problem_list.index(problem) + 0.2, y1=1.5,
            line_color='black'
        )
        fig_arrow.add_annotation(
            x=len(process_list) + problem_list.index(problem), y=1,
            text=problem,
            showarrow=False
        )

    # アローの追加
    for process in process_list:
        for problem in problem_list:
            if relation_dict[process][problem] > 0:
                fig_arrow.add_annotation(
                    x=process_list.index(process), y=1.2,
                    ax=len(process_list) + problem_list.index(problem), ay=1.2,
                    xref='x', yref='y', axref='x', ayref='y',
                    text=relation_dict[process][problem],
                    showarrow=True, arrowhead=1
                )

    fig_arrow.update_layout(title="アローダイヤグラム")
    fig_arrow.show()

# 不具合記録データからアローダイヤグラムを作成
create_arrow_diagram(df_bugu)

# ... (既存のコードは省略) ...
```

```python
# prompt: 不具合記録は文章で記録します。PDPC法を扱うライブラリがあれば例題を見せてください。新QC7つ道具の言語データの処理には大規模言語モデル（LLM）を使用して文章のテキストマイニングを行ってください。plotlyで作図・作表してください。

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from collections import defaultdict

# ... (既存のコードは省略) ...


# PDPC法の簡易的な実装（例）
def create_pdpc_diagram(df_bugu):
    """
    PDPC図を作成する関数
    """
    process_list = sorted(list(set(df_bugu['工程'])))
    problem_list = sorted(list(set(df_bugu['不良品の種類'])))

    # ノード間の関係性を表す辞書
    relation_dict = {}
    for process in process_list:
        relation_dict[process] = {}
        for problem in problem_list:
            relation_dict[process][problem] = 0

    for index, row in df_bugu.iterrows():
        relation_dict[row['工程']][row['不良品の種類']] += row['発生数']

    # グラフの作成
    fig_pdpc = go.Figure()

    # 工程ノード
    for i, process in enumerate(process_list):
        fig_pdpc.add_shape(
            type='rect',
            xref='x', yref='y',
            x0=i * 2, y0=0,
            x1=i * 2 + 1, y1=1,
            line_color='black'
        )
        fig_pdpc.add_annotation(
            x=i * 2 + 0.5, y=0.5,
            text=process,
            showarrow=False
        )

    # 不良品の種類ノード
    for i, problem in enumerate(problem_list):
        fig_pdpc.add_shape(
            type='rect',
            xref='x', yref='y',
            x0=len(process_list) * 2 + i * 2, y0=0,
            x1=len(process_list) * 2 + i * 2 + 1, y1=1,
            line_color='black'
        )
        fig_pdpc.add_annotation(
            x=len(process_list) * 2 + i * 2 + 0.5, y=0.5,
            text=problem,
            showarrow=False
        )

    # アローの追加
    for process in process_list:
        for problem in problem_list:
            if relation_dict[process][problem] > 0:
                fig_pdpc.add_annotation(
                    x=process_list.index(process) * 2 + 0.5, y=0.8,
                    ax=len(process_list) * 2 + problem_list.index(problem) * 2 + 0.5, ay=0.8,
                    xref='x', yref='y', axref='x', ayref='y',
                    text=relation_dict[process][problem],
                    showarrow=True, arrowhead=1
                )

    fig_pdpc.update_layout(title="PDPC図", xaxis_visible=False, yaxis_visible=False)
    fig_pdpc.show()

# 不具合記録データからPDPC図を作成
create_pdpc_diagram(df_bugu)

# ... (既存のコードは省略) ...
```
