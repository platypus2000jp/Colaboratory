```python
# prompt: 魚の写真をアップロードします。この魚の魚拓を作成し、左に戻画像、右に魚拓を並べて表示してください。そして魚拓の画像はダウンロードして、名前を付けて保存します。

from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np

# 画像のアップロード（Google Colabの場合）
from google.colab import files
uploaded = files.upload()

# アップロードされた画像ファイル名を取得
image_name = list(uploaded.keys())[0]

# 画像を開く
img = Image.open(image_name)

# 魚拓作成
# グレースケール変換
gray_img = ImageOps.grayscale(img)

# 反転
inverted_img = ImageOps.invert(gray_img)


# 画像表示
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(inverted_img, cmap='gray')
axes[1].set_title('Gyotaku')
axes[1].axis('off')

plt.tight_layout()
plt.show()


# 魚拓画像のダウンロード
# ファイル名を設定
output_filename = "gyotaku_" + image_name

# 画像を保存
inverted_img.save(output_filename)


# ダウンロード処理
files.download(output_filename)
```
